```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("3.csv")
country = data['Country']
xs = data['Recov.']
ys = data['Death']

plt.figure(figsize = (10,10))
plt.scatter(xs,ys)
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Death', fontsize=15)
plt.xlabel('Recoveries', fontsize=15)
plt.title('Ratio of the number of dead from Covid 19 to recovered, by country, log scale')
plt.suptitle('15-Jun-2020')
for x,y,country in zip(xs,ys, country):

    label = str(country)

    # this method is called for each point
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center',
                fontsize=10) # horizontal alignment can be left, right or center
line = plt.plot(np.unique(xs), np.poly1d(np.polyfit(xs, ys, 1))(np.unique(xs)))
```
