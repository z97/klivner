```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel (r'2.xlsx')
df2 = df[df.percentage != '-' ]
df3 = df2[df2.percentage != '..' ]
x = df3['percentage'].astype(float)
y = df3['salary'].astype(float)
s = df3['population']
t = df3['city']
```

```python
x_list = x.tolist()
y_list = y.tolist()
t_list = t.tolist()
s_list = s.tolist()
```

```python
reversed_strings = [x[::-1] for x in t_list]
```

```python
types = reversed_strings
x_coords = x_list
y_coords = y_list
s_coords = s_list
plt.figure(figsize=(15,11))
for i,type in enumerate(types):
    x = x_coords[i]
    y = y_coords[i]
    s = s_coords[i]
    plt.scatter(x, y, s, color='lightgray')
    plt.text(x, y, type, fontsize=12)
plt.title("scatterplot of average salary and percentage of repatriates")
plt.xlabel("percentage of repatriates")
plt.ylabel("average salary")
plt.show()
```
