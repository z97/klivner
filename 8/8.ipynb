```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_json (r'5.json')
```

```python
idx = pd.date_range(start='10/1/1994', end='11/1/2019', freq='MS')    
s = pd.Series(df['rain_ttl'].values, index=df['time_obs'])
```

```python
s.index = pd.DatetimeIndex(s.index)
s = s.reindex(idx, fill_value=0)
```

```python
df2=s.to_frame(name="rain_ttl")
df3 = df2.round(0)
```

```python
df3['month'] = df3.index.month
df3['year'] = df3.index.year
```

```python
index= df3['year']
cols = df3['month']
```

```python
sns.set()
flights = df3.pivot("year", "month", "rain_ttl")
f, ax = plt.subplots(figsize=(15, 20))
sns.heatmap(flights, linewidths=.5, annot=True, fmt='.0f', cmap="YlGnBu")
ax.set_title('History of monthly rainfall in Rishon Lezion')
```
